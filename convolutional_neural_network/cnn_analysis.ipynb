```python
import matplotlib.pyplot as plt
import PIL
import torchvision.transforms as transforms

import torch
import torch.nn as nn
```

```python
img = PIL.Image.open("./cat.jpeg")
```

```python
tf_tt = transforms.ToTensor()
tf_tpi = transforms.ToPILImage()

img_t = tf_tt(img)
```

```python
print("img_t.shape : ",img_t.shape)
```

```python
conv_v1 = nn.Conv2d(
    in_channels=3,
    out_channels=16,
    kernel_size=32,
    padding=0
)
img_t_conved = conv_v1(img_t)
```

```python
fig, axs = plt.subplots(4, 4,figsize=(16,16))
for x_idx in range(4):
    for y_idx in range(4):
        idx = x_idx * 4 + y_idx
        tmp = img_t_conved[idx]
        tmp_img = tf_tpi(tmp)
        axs[x_idx][y_idx].imshow(tmp_img)
```

```python
max_pool_v1 = nn.MaxPool2d(
    kernel_size=8,
    stride=2
)
img_t_conved_max_pooled = max_pool_v1(img_t_conved)
```

```python
fig, axs = plt.subplots(4, 4,figsize=(16,16))
for x_idx in range(4):
    for y_idx in range(4):
        idx = x_idx * 4 + y_idx
        tmp = img_t_conved_max_pooled[idx]
        tmp_img = tf_tpi(tmp)
        axs[x_idx][y_idx].imshow(tmp_img)
```
